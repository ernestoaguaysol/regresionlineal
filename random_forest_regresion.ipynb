```python
# Importando las librerías
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importando el dataset
dataset = pd.read_csv('dataset_salarios_posicion.csv')
X = dataset.iloc[:, 1:-1].values # idem que [:, 1:2]
y = dataset.iloc[:, -1].values # idem que [:, 2]
```

```python
# Importar el regresor
from sklearn.ensemble import RandomForestRegressor

# Crear el regresor
# Parámetro random_state: semilla del generador random. 
regresor = RandomForestRegressor(n_estimators=10, random_state = 0)
regresor.fit(X, y)
```

```python
# Ploteo el dataset.
plt.scatter(X, y, color = 'red') 
 
# Ploteo de los resultados de predecir con random forest
X_grid = np.arange(min(X), max(X), 0.01) # Datos para la predicción, entre el mínimo y el máximo de X, step de 0.01
X_grid = X_grid.reshape((len(X_grid), 1)) # 
plt.plot(X_grid, regresor.predict(X_grid), color = 'blue')
 
plt.title('Random Forests (10 estimadores) - Regresión')
plt.xlabel('Nivel de posición')
plt.ylabel('Salario')
plt.show()
```

```python
######################################################################################################

# Crear el regresor
from sklearn.tree import DecisionTreeRegressor
```

```python
# Crear el regresor
# Parámetro criterion: el criterio del regresor. Default 'mce' = promedio de errores al cuadrado.
# Parámetro random_state: semilla del generador random. Default 'None' = lo que sale de np.random
regresor = DecisionTreeRegressor(random_state = 0)
regresor.fit(X, y)
```
