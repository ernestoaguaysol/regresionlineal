```python
# Importamos las librer√≠as necesarias
from sklearn.datasets import load_iris
from sklearn import tree
```

```python
# Cargamos en X e y los datos
X, y = load_iris(return_X_y=True)
```

```python
# Creamos el clasificador
arbol_clasificacion = tree.DecisionTreeClassifier()
```

```python
# Entrenamos el clasificador con los datos de entrenamiento
arbol_clasificacion = arbol_clasificacion.fit(X, y)
```

```python
# Podemos imprimir el √°rbol con informaci√≥n de caracter√≠sticas y clases utilizando los datos de iris y el clasificador entrenado

datos_iris = load_iris()
tree.plot_tree(	arbol_clasificacion, 
                feature_names = datos_iris.feature_names, 
                class_names=datos_iris.target_names)
```

```python
# Para realizar una predicci√≥n, se llama a la funci√≥n
arbol_clasificacion.predict()
# Pasando como par√°metro los valores de la observaci√≥n para la cual necesitamos predecir su clase.
```

```python
#######################################################################################################
# Ejemplo de clasificaci√≥n dataset Iris - largo y ancho del s√©palo
# Se muestra ahora un ejemplo tomando las dos primeras caracter√≠sticas del dataset:
X, y = load_iris(return_X_y=True)
X2 = X[:, 0:2] # Tomamos las dos primeras caracter√≠sticas
arbol_clasificacion2 = tree.DecisionTreeClassifier()
arbol_clasificacion2 = arbol_clasificacion2.fit(X2, y)
```

```python
# Vemos que el a≈ïbol generado contiene muchos nodos para el tama√±o del dataset, esto significa que el √°rbol est√° generado muy a medida del conjunto de datos, indicando sobreentrenamiento.


# Como son dos caracter√≠sticas, es posible dibujar en 2D la distribuci√≥n de los datos y la superficie de decisi√≥n generada por el √°rbol:

from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
```

```python
# Graficar
ax = plot_decision_regions(X2, y, clf=arbol_clasificacion2, legend=2)
plt.xlabel('Largo del s√©palo [cm]')
plt.ylabel('Ancho del s√©palo [cm]')
 
handles, labels = ax.get_legend_handles_labels()

ax.legend(handles, 
          ['setosa', 'versicolor', 'virginica'], 
           framealpha=0.3, scatterpoints=1)
plt.show()
```
