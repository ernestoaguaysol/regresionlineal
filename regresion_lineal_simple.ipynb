```python
# Importar las librerías
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('dataset_hs_jugadas_puntaje.csv')
# Todas las filas y todas las columnas exepto la última
X = dataset.iloc[:, :-1].values
# Todas las filas y la última columna (en este caso sabemos que tenemos dos columnas 0 y 1)
y = dataset.iloc[:, 1].values
```

```python
from sklearn.linear_model import LinearRegression
regresor = LinearRegression()
regresor.fit(X, y)
```

```python
y_predicciones = regresor.predict(X)
```

```python
# para dibujar grafico de disperción
plt.scatter(X, y, color='blue')
# para dibujar una linea
plt.plot(X, y_predicciones, color='green')
plt.show()
```
